= (4ax^2+a)+3+(1/(1+e^-z))  
σ(z)= 1/1+e^-z

```python
x, y, z, a, b = sp.symbols('x y z a b')
```

```python
# Define the function f(x, z, a, b)
f = (4*a*x**2 + a) + 3 + sp.exp(z)/(1 + sp.exp(z)) + (sp.exp(b)/(1 + sp.exp(b)))**2

# Calculate the partial derivatives
df_dx = sp.diff(f, x)
df_dz = sp.diff(f, z)
df_da = sp.diff(f, a)
df_db = sp.diff(f, b)

# Print the partial derivatives
print("Partial derivative of f with respect to x:")
sp.pprint(df_dx)

print("\nPartial derivative of f with respect to z:")
sp.pprint(df_dz)

print("\nPartial derivative of f with respect to a:")
sp.pprint(df_da)

print("\nPartial derivative of f with respect to b:")
sp.pprint(df_db)
```
